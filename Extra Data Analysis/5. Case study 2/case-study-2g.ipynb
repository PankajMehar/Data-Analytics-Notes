# Exploring with Visuals
Use `clean_08.csv` and `clean_18.csv`

```python
import pandas as pd
import numpy as np

# load datasets

df_08 = pd.read_csv('clean_08.csv')
df_18 = pd.read_csv('clean_18.csv')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set_style('darkgrid')
sns.set(rc={'figure.figsize':(10,8)}) 
```

## Distributions of greenhouse gas score in 2008 and 2018

```python
df_08['greenhouse_gas_score'].hist();
plt.xlabel('greenhouse_gas_score');
plt.ylabel('Count');
plt.title('Distributions of Green house gas score for 2008');
```

```python
df_18['greenhouse_gas_score'].hist();
plt.xlabel('greenhouse_gas_score');
plt.ylabel('Count');
plt.title('Distributions of Green house gas score for 2018');
```

## distribution of combined mpg changed from 2008 to 2018

```python
df_08['cmb_mpg'].plot(kind='hist');
plt.xlabel('combined mpg');
plt.ylabel('Count');
plt.title('Distributions of combined mpg for 2008');
```

```python
df_18['cmb_mpg'].plot(kind='hist');
plt.xlabel('combined mpg');
plt.ylabel('Count');
plt.title('Distributions of combined mpg for 2018');
```

__Observe__ : Become much more skewed to right.

## Describe the correlation between displacement and combined mpg.

```python
sns.lmplot(x = 'displ', y = 'cmb_mpg', data = df_08);
plt.ylabel('combined mpg');
plt.xlabel('Displacement');
plt.title('Correlation between combined mpg and Displacement for 2008');
```
